```python
poblacion=poblacion.iloc[:53]
```

```python
def extraer_enero(x):
    if isinstance(x, str):
        if re.search("enero", x, re.IGNORECASE):
            match_enero=re.search(r"\d{4}$", x)
            if match_enero:
                return match_enero.group()
    return None 
```

```python
poblacion["Periodo"]=poblacion["Periodo"].apply(extraer_enero)

poblacion=poblacion[poblacion["Periodo"].notna()]

poblacion=poblacion.rename(columns={"Total": "Población residente total"})

print(poblacion)
```

```python
poblacion.loc[len(poblacion)]=['2024', poblacion_1enero2024["Total"][0]]
poblacion=poblacion.sort_values(by='Periodo', ascending=False)
```

```python
poblacion.iloc[:,0]=poblacion.iloc[:,0].astype(int)-1
```

```python
poblacion.iloc[:,0]=pd.to_numeric(poblacion.iloc[:,0], errors='coerce')
poblacion.iloc[:,1]=poblacion.iloc[:,1].str.replace('.', '')
poblacion.iloc[:,1]=poblacion.iloc[:,1].str.replace(',', '.').astype(float)
poblacion.iloc[:,1]=pd.to_numeric(poblacion.iloc[:,1], errors='coerce')
```

```python
# Poblacion extranjera total

poblacion_extranjera=pd.read_csv("/home/pablo/Documentos/CSVs TFG/Poblacion_extranjera_total.csv", sep=";")
poblacion_extranjera=poblacion_extranjera[(poblacion_extranjera["Nacionalidad"] == "TOTAL EXTRANJEROS") & 
                                           (poblacion_extranjera["Total Nacional"] == "Total Nacional") &
                                           (poblacion_extranjera["Comunidades y Ciudades Autónomas"].isna()) &
                                           (poblacion_extranjera["Sexo"] == "Ambos sexos")
                                           ]
                                           
poblacion_extranjera=poblacion_extranjera.drop(columns=["Nacionalidad", "Total Nacional", 
                                                         "Comunidades y Ciudades Autónomas", "Sexo"])
                                                         
poblacion_extranjera=poblacion_extranjera.rename(columns={"Total": "Población extranjera total",
                                                           "Año": "Periodo"})
```

```python
poblacion_extranjera=poblacion_extranjera.iloc[:14]
```

```python
poblacion_extranjera.iloc[:,0]=pd.to_numeric(poblacion_extranjera.iloc[:,0], errors='coerce')

poblacion_extranjera.iloc[:,1]=poblacion_extranjera.iloc[:,1].str.replace('.', '').astype(float)
poblacion_extranjera.iloc[:,1]=pd.to_numeric(poblacion_extranjera.iloc[:,1], errors='coerce')
```

arreglo

```python
años_a_extrapolar=[2023]

df=pd.DataFrame()

df['Tasa_Crecimiento']=poblacion_extranjera.iloc[:,1].pct_change()

tasa_ultimo_año=df['Tasa_Crecimiento'].iloc[1]

valores_extrapolados_tasa=[]
ultimo_valor_conocido=poblacion_extranjera.iloc[:,1].iloc[0]
for _ in años_a_extrapolar:
    siguiente_valor=ultimo_valor_conocido * (1 + -tasa_ultimo_año)
    valores_extrapolados_tasa.append(siguiente_valor)
    ultimo_valor_conocido=siguiente_valor

df_extrapolado_tasa=pd.DataFrame({'Periodo': años_a_extrapolar, 'Población extranjera total': valores_extrapolados_tasa})

df_resultados_tasa=pd.concat([poblacion_extranjera,df_extrapolado_tasa], ignore_index=True)

poblacion_extranjera=df_resultados_tasa

poblacion_extranjera=poblacion_extranjera.sort_values(by=poblacion_extranjera.columns[0], ascending=False)
```
